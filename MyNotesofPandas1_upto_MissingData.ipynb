## MissingData

```python
import numpy as np
import pandas as pd
```

Finding Missing Data

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
#np.nan is not a number i.e null value
df
```

```python
df.isna()# for true and false jha jha nan hai wha wha true
```

```python
df.isna().sum()# sum lagaye hai count karne ke liye nan
```

```python
df.isna().any()# ye dikahata hai ki sab me nan hai ya koi missing hai sara true hai mtlb sab me nan value hai
```

Removing Missing Data

works on the basis of rows

```python
df
```

```python
df.dropna()#dropna se sare nan value wale rows delete ho jayega sab me nan hai tu sare delete ho jayenge
#but maine abhi dubara jake data me ek nan ko hta ke 1 kiya hai isko sahi karne ke liye
```

```python
df
```

```python
df.dropna(thresh=2)#thresh=2 ka mtlb hai atleast 2 non null value
```

```python
df.dropna(thresh=3)
```

```python
df.dropna(thresh=4)
```

Filling the missing Data

```python
df
```

```python
df.fillna(0)#sare nan 0 se fill ho jayenge
```

**Remembering trick**

remove ke liye (.dropna)

fill ke liye (.fillna)

search ke liye (.isna)

```python
values = {'A':0,'B':100,"C":300,'D':400}
df.fillna(value=values)#fillna column wise and dropna rowwise
```

```python
df.fillna(df.mean())# sabke mean se badhenge not that useful in reallife
```
