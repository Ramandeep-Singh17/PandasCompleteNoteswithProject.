## 🧠 1. Pandas kya hai?

```python
import pandas as pd
import numpy as np
```

### ✅ Kya?
Pandas ek Python library hai jo data ko manipulate, clean aur analyze karne ke kaam aati hai. Isme do main data structures hote hain:
- Series: 1D labeled data (list jaisa)
- DataFrame: 2D table jaisa data (Excel sheet jaisa)

### ✅ Kyun?
- Structured data ko aasani se handle kar sakte hain
- CSV, Excel, JSON jaisi files ko load kar sakte hain
- Fast, flexible aur readable syntax

### ✅ Kab Use Karein?
- Jab table jaise data ke saath kaam ho (rows & columns)
- Data cleaning, transformation, aur analysis ke time

### ✅ Real-Life Example
- Sales report Excel file ko clean karke analysis karna ho.

## 📦 2. Series & DataFrame

```python
# Series – 1D labeled data
ds = pd.Series([10, 20, 30])
print(ds)

# DataFrame – 2D table
data = {'Name': ['Alice', 'Bob'], 'Age': [25, 30]}
df = pd.DataFrame(data)
print(df)
```

```python
pd.read_csv("file.csv")       # CSV file load karna
pd.read_excel("file.xlsx")    # Excel file load
pd.read_json("file.json")     # JSON file load
```

## 🛠️ 4. Basic Operations

```python
df.head()       # Pehli 5 rows
df.tail()       # Aakhri 5 rows
df.shape        # Rows aur columns ka size
df.info()       # Column type aur missing values
df.describe()   # Numeric stats summary
```

## 🎯 5. Data Select & Filter

```python
df['Age']                    # Ek column
df[['Name', 'Age']]          # Multiple columns
df.loc[2]                    # Row label se
df.iloc[2]                   # Row position se
df[df['Age'] > 25]           # Filter lagana
```

---

## ❌ 6. Missing Data

```python
df.isnull().sum()                      # Kitni null values hain
df.dropna()                            # Null wali rows hatao
df.fillna(0)                           # Null ko 0 se bharo
df['col'].replace('Unknown', np.nan)  # 'Unknown' ko NaN bana do
```

---

## 🪝 7. Data Clean & Type Change

```python
df['Age'] = df['Age'].astype(int)     # Integer me convert karna
df['Name'] = df['Name'].str.strip()   # Extra space hataana
df['Name'] = df['Name'].str.lower()   # Lowercase me karna
```

## 🗖️ 14. DateTime Columns

```python
# Date column ko proper datetime object me convert kar rahe hain
# Taaki hum year, month, day jaise attributes nikal sakein

df['Date'] = pd.to_datetime(df['Date'])  # String to datetime

df['Year'] = df['Date'].dt.year          # Year extract kar rahe hain
df['Month'] = df['Date'].dt.month        # Month extract kar rahe hain
```
